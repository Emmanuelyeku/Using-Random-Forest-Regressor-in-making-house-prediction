```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
training_dataset = pd.read_csv("msc_training_dataset.csv")
```

```python
training_dataset.head()
```

```python
training_dataset.describe()
```

```python
X = training_dataset.drop('price', axis=1)
y = training_dataset['price']
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
importances = model.feature_importances_
```

```python
for feature, importance in zip(X.columns, importances):
    print(f'{feature}: {importance}')
```

```python
sorted_indices = np.argsort(importances)[::-1]
sorted_importances = [importances[i] for i in sorted_indices]
sorted_features = [X.columns[i] for i in sorted_indices]
plt.bar(range(len(sorted_importances)), sorted_importances)
plt.xticks(range(len(sorted_importances)), sorted_features, rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances (Sorted)')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print('Mean Squared Error:', mse)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print('R square:', r2)
```

```python
from scipy.stats import ttest_ind
price = test_dataset['price']
predicted_prices = test_dataset['predicted_prices']

t_stat, p_value = ttest_ind(price, predicted_prices)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```
